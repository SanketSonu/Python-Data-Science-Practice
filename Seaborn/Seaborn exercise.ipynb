```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')   #Loading inbuild dataset.
tips.head()
```

# Distplot:

```python
sns.displot(tips['total_bill'], bins = 30, color='green', kde=True)
```

# Jointplot:

```python
sns.jointplot(x='total_bill', y='tip', data = tips, color='orange')
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, color='orange', kind = 'hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, color='red', kind = 'reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, color='blue', kind = 'kde')
```

# Pairplot:

```python
sns.pairplot(tips, palette='coolwarm', hue='sex')
```

# Rugplot:

```python
sns.rugplot(tips['total_bill'], color= 'red')
```

# KDE Plot:

```python
sns.kdeplot(tips['total_bill'], color='red', lw=2, ls = '--')
```

# Barplot:

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

# Countplot:

```python
sns.countplot(x='sex', data=tips)
```

# Boxplot:

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
# hue = 'smoker' to split box into smokers category. This shows total_bill of male and female who are smokers and non-smokers.
sns.boxplot(x='sex', y='total_bill', data=tips, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# hue = 'sex' to split box into sex category. This shows total_bill on each day where customers are male and female.
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# hue = 'smoker' to split box into smokers category. This shows total_bill on each day where customers are smokers and non-smokers.
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

# Violinplot:

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# hue to split these plots into sex category.
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# spit=True to combine hue=sex of these violin plots.
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

# Stripplot:

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
#hue='sex' to split between male and female and jitter=True to provide noise or more clearance to the data.
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
```

# Swarmplot:

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
#This is much better than just using Swarmplot because Swarmplot is not good for very large dataset. 
#As points stack on top od eachother.

sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```


# Factorplot:

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# To convert dataset into matrix to plot Matrix plots.
tc = tips.corr()
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, lw=1, linecolor='white')
```

# Clustermap:

```python
sns.clustermap(fp, standard_scale=1)
```

```python
#importing a new iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

# Pairplot:

```python
sns.pairplot(iris)
```

```python
#If we want to customize this pairplot, we can do it by first creating an empty grid
g = sns.PairGrid(iris)
```

```python
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# FaceGrid:

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g.map(plt.scatter, 'total_bill', 'tip')
```
