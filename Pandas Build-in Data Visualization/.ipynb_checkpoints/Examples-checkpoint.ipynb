```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

###### We can use any 1 method of 3 mentioned. All are same and it support matplotlib stylings.

# Hiostogram:

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist(bins=30)
```

# Areaplot:

```python
df2.plot.area(alpha=0.4, figsize=(12,6))
```

# Boxplot:

```python
df2.plot.box()
```

# Barplot:

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

# Scatterplot:

```python
df1.plot.scatter(x='A', y='B')
```

# Hexbin:

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.plot.hexbin(x='a', y='b', gridsize=20, cmap='coolwarm')
```

# KDE plot:

```python
df2.plot.kde()
```

```python
#Both KDE and DENSITY are same.
df2['a'].plot.density()
```
