```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

observations = 1000

xs = np.random.uniform(low = -10, high = 10, size = (observations,1))
zs = np.random.uniform(-10, 10, (observations,1))

inputs = np.column_stack((xs,zs))
print("Size of inputs", inputs.shape)

noise = np.random.uniform(-1, 1, (observations,1))
targets = 2*xs - 3*zs + 5 + noise
print("Size of targets", targets.shape)

#Radius of the range to initialize Weights and Biases
init_range = 0.1
weights = np.random.uniform(-init_range, init_range, size = (2,1))
biases = np.random.uniform(-init_range, init_range, size = 1)
print("Weights are:", weights)
print("Bias is:", biases)

learning_rate = 0.02

print("\n")
print("Printing Loss after each iteration: \n")
for i in range(100):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
    loss = np.sum(deltas ** 2) / 2 / observations
    print(loss)
    deltas_scaled = deltas / observations
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)


print("Best Weights are: \n",weights)
print("Best Bias is: \n",biases)

print("Below is the graph which is very close or equal to 45°, which indicates the closeness of our 'Outputs' with the 'Targets'.")

plt.plot(outputs,targets)
plt.xlabel('Outputs', fontsize = 15)
plt.ylabel('Targets', fontsize = 15)
plt.show() 
```
