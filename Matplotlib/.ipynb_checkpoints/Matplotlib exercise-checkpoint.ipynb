```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,21)
y = x ** 2
```

```python
print('X Values are:',x)
print('Y Values are:',y)
```

# Functional method:

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)    #(Row, Column, Index)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')

plt.subplot(1,2,2)    #(Row, Column, Index)
plt.plot(y,x)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')

plt.tight_layout()    #To provide space between overlapping subplots.
```

# Object Oriented method:

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])
ax3 = fig.add_axes([0.75,0.15,0.2,0.3])

ax1.plot(x,y, color = 'red', lw=3, marker = 'o', markerfacecolor = 'yellow')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Title')

ax2.plot(y,x, color = 'blue', lw=3, alpha = 0.5)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Title')

ax3.plot(x,y, color = 'purple', label = 'X', lw=2, ls='--')
ax3.plot(y,x, color = 'green', label = 'Y', lw=2, ls='--')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_title('Title')
ax3.legend()
```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x,y)
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('Title')

ax[1].plot(y,x)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('Title')

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'X', lw = 2, marker = 'o', markersize = 5, markerfacecolor = 'red')
ax.plot(y,x, label = 'Y', lw = 2, marker = 'o', markersize = 5, markerfacecolor = 'red')
ax.set_xlim([0,2])
ax.set_ylim([0,2])
ax.legend(loc = 2)
```
